```python
# Loading dataset
df = pd.read_csv("body.csv", sep=';')

sizes = df[["Shoulder", "Chest", "Waist","Navel","Hip","Thigh","Bicep","Forearm", "Knee","Calf","Ankle","Wrist","Weight","Height","Gender"]].dropna(0, 'any')

Height=df.Height
Shoulder=df.Shoulder
    
```

```python
# Plotting unique pairs of (length, width)
def rect_range(rangex, rangey):
    """ Creates a list of abcissas and ordinates,
        to plot a rectangle.
    """
    minx, maxx = rangex
    miny, maxy = rangey
    return [minx, maxx, maxx, minx, minx],\
           [miny, miny, maxy, maxy, miny]

    
#green_zone = rect_range((3000, 6500), (1400, 2400))
#grey_zone_a = rect_range((3000, 4100), (2050, 2400))
#grey_zone_b = rect_range((5250, 6500), (1400, 1650))

plt.figure(figsize=(12, 12))
plt.plot([Height[i] for i in range(len(Height))], [Shoulder[i] for i in range(len(Shoulder))],'o', label='Unique', c='blue')

#plt.plot(green_zone[0], green_zone[1], label='Default range', c='green')
#plt.plot(grey_zone_a[0], grey_zone_a[1],c='grey')
#plt.plot(grey_zone_b[0], grey_zone_b[1],c='grey')

plt.title("Pedestrian Height by shoulder grith")
plt.xlabel("Height (cm)")
plt.ylabel("width (mm)")
#plt.xlim(2000, 7000)
#plt.ylim(800, 2500)
plt.legend(loc='best')
plt.show()
```
