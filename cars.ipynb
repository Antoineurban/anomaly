# Cars dimensions

```python
# Imports
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Loading dataset
df = pd.read_csv("CQA_Premium.csv")
print("shape:", df.shape)
```

```python
# Plotting height and width
plt.figure(figsize=(12, 7))
plt.plot(df['model_width_mm'], df['model_height_mm'], 'o')
plt.title("Car height by width")
plt.ylim([0, 4000])
plt.xlim([0, 10000])
plt.xlabel("width (mm)")
plt.ylabel("height (mm)")
plt.show()
```
